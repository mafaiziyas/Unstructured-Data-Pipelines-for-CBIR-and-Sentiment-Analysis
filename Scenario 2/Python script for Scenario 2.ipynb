```python
#Data ingestion

import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/Data Engineering CW2/Scenario 2/Raw Data/tweets.csv")

print(df.head())
print(df.shape[0])
```

```python
#Task b
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    #Text normalisation (lowercasing)
    text = str(text).lower()

    #Tokenisation
    tokens = word_tokenize(text)

    #Removal of punctuation and stop words
    tokens = [
        word for word in tokens
        if word not in string.punctuation and word not in stop_words
    ]

    #Rejoins the cleaned words
    return " ".join(tokens)

#Applies preprocessing to the dataset
df['cleaned_text'] = df['message to examine'].apply(preprocess_text)
#Display
df.head()
```

```python
#Task c.i

from sklearn.feature_extraction.text import CountVectorizer

# Initialise Bag of Words vectorizer
bow_vectorizer = CountVectorizer()

# Fit and transform cleaned text
X_bow = bow_vectorizer.fit_transform(df['cleaned_text'])

# Convert to DataFrame for readability
bow_df = pd.DataFrame(
    X_bow.toarray(),
    columns=bow_vectorizer.get_feature_names_out()
)

#Display
print("Bag of Words shape:", bow_df.shape)
bow_df.head()
```

```python
#Task c.ii

from sklearn.feature_extraction.text import TfidfVectorizer

# Initialise TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform cleaned text
X_tfidf = tfidf_vectorizer.fit_transform(df['cleaned_text'])

# Convert to DataFrame
tfidf_df = pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

print("TF-IDF shape:", tfidf_df.shape)
tfidf_df.head()
```

```python
#Task d

from textblob import TextBlob
import json

# 1. Automatic Sentiment Labeling
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

df['sentiment'] = df['cleaned_text'].apply(get_sentiment)

# 2. Creating the Metadata
metadata = {
    "project_name": "Text Sentiment Analysis",
    "vectorization_methods": {
        "method_1": {
            "type": "Bag of Words (CountVectorizer)",
            "description": "Converts text to a matrix of token counts.",
            "vocab_size": len(bow_vectorizer.get_feature_names_out())
        },
        "method_2": {
            "type": "TF-IDF (TfidfVectorizer)",
            "description": "Weights tokens by frequency and inverse document frequency.",
            "vocab_size": len(tfidf_vectorizer.get_feature_names_out())
        }
    },
    "data_summary": {
        "total_documents": len(df),
        "sentiment_counts": df['sentiment'].value_counts().to_dict()
    }
}

# Save metadata to a JSON file
with open('metadata.json', 'w') as f:
    json.dump(metadata, f, indent=4)

print("Metadata and Labelling complete.")
print(df[['message to examine', 'sentiment']].head())
```
